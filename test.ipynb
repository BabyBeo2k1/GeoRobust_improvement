```python
import torch.nn.functional as F
class AffineTransf_grad(nn.Module):

    def __init__(self, theta):
        super(AffineTransf_grad, self).__init__()
        self.theta = theta.view(-1,2,3)
        
    def forward(self, img):
        batch_size, channel, height, width = img.size()

        grid = F.affine_grid(self.theta, torch.Size((
            batch_size, channel, height, width)))
        transfored_img = F.grid_sample(img, grid)
        return transfored_img

img=torch.rand(3,2,2)
print(img)
imgs=img.repeat(5,1,1).view((5,3,2,2))
print(imgs.shape)
val=[[[1.,0.,0.],[0.,1.,0.]]]
theta=torch.tensor(val).repeat(5,1,1)
print(theta.size())
aff=AffineTransf_grad(theta)
out=aff(imgs)
print(out.size())
```
